```python
from bw_recipe_2016.chemidplus import ChemIDPlus, Multiple, Missing
```

```python
c = ChemIDPlus()
```

```python
from bw_recipe_2016 import get_biosphere_database
```

```python
biosphere = get_biosphere_database()
```

```python
biosphere_cas = {o["name"]: o.get("CAS number") for o in biosphere}
```

```python
biosphere_names = sorted({o['name']: o for o in biosphere})
```

```python
import pyprind
```

```python
multiple = []
missing = []

for o in pyprind.prog_bar(biosphere_names):
    try:
        c.add_master_term(o, biosphere_cas.get(o))
    except Multiple:
        multiple.append(o)
    except Missing:
        missing.append(o)
```
